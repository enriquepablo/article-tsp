## Sets and ordering.

Here we see how we will be representing instances of the SP and the TSP in Python,
and how we will represent the symmetric group and its action on these instances of the problems.

First we import numpy and pyplot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

A couple of functions to generate random instances of both problems:

```python
def random_sp(n):
    return np.random.random(n)

def random_tsp(n):
    return np.squeeze(np.random.random((n, 2)).view(complex))
```

So we use numpy arrays for our instances of the problem,
which means that we have both the set part and the ordering part
of the formulation of each instance of each problem
encoded in the same Python object.

```python
sp1 = random_sp(30)

tsp1 = random_tsp(30)
```

We can plot these instances

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
axes[0].plot(sp1)
axes[1].plot(tsp1.real, tsp1.imag)
fig.tight_layout()
```

Obviously for solved instances, the plot for the SP will look (and be) monotonic,
and for the TSP will look short and without crossings.

We will represent the symmetric groups as permutation matrices.
So for example a transposition $(2 3)$ in $\mathcal{S}_5$ would look like:

```python
eye = np.eye(5, dtype=np.uint8)

t23 = np.eye(5, dtype=np.uint8)
t23[2] = eye[3]
t23[3] = eye[2]

print(t23)
```

Now if we have an instance of the SP with 5 elements

```python
sp2 = np.arange(5)
np.random.shuffle(sp2)

print(sp2)
```

we can obtain a new ordering of the instance acting on it with the permutation matrix `t23`

```python
print(sp2 @ t23)
```
